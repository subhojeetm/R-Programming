<strong>Fuzzy Match: </strong>If we do not know the exact match, then we can search the documentation using Fuzzy match.
<br>Fuzzy match is performed using ?? (double question mark symbol)
<br>If you are searching for a single word, no need to include the word inside quotes. However, you can.
<br>If there are multiple words that you are trying to search, then it has to be included in quotation mark. Single or Double both works.

```r
??regression
```

I got the results in a different tab.

```r
??"linear regression"
```

apropos() function gives you a list of built in functions as well as functions defined in various packages matching the string passed to apropos() function. Finally it also shows the variable matching the name. See yello highlighted example above.

* apropos() function also works with regular expressions

```r
variable_a <- 20
variable_b <- 30

variable_result <- variable_a + variable_b

variable_result
```

```r
apropos("^var")
```

* in the example above ^var means -> shows all built in function names and variable names starting with "var".
